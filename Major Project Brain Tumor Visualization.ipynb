```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Define paths to images
image_paths = {
    'glioma_axial': "F:/Final Project/brain/glioma/Tr-gl_0053.jpg",
    'glioma_coronal': "F:/Final Project/brain/glioma/Tr-gl_0037.jpg",
    'glioma_sagittal': "F:/Final Project/brain/glioma/Tr-gl_0045.jpg",
    'meningioma_axial': "F:/Final Project/brain/meningioma/Tr-me_0033.jpg",
    'meningioma_coronal': "F:/Final Project/brain/meningioma/Tr-me_0034.jpg",
    'meningioma_sagittal': "F:/Final Project/brain/meningioma/Tr-me_0206.jpg",
    'pituitary_axial': "F:/Final Project/brain/pituitary/Tr-pi_0023.jpg",
    'pituitary_coronal': "F:/Final Project/brain/pituitary/Tr-pi_0028.jpg",
    'pituitary_sagittal': "F:/Final Project/brain/pituitary/Tr-pi_0019.jpg",
    'notumor_axial': "F:/Final Project/brain/notumor/Tr-no_0011.jpg",
    'notumor_coronal': "F:/Final Project/brain/notumor/Tr-no_0275.jpg",
    'notumor_sagittal': "F:/Final Project/brain/notumor/Tr-no_0278.jpg"
}

# Define class names
class_names = {
    'glioma': 'Glioma',
    'meningioma': 'Meningioma',
    'pituitary': 'Pituitary',
    'notumor': 'No Tumor'
}

# Display sample images
fig, axes = plt.subplots(4, 3, figsize=(7, 7))
for i, (class_name, class_label) in enumerate(class_names.items()):
    for j, view in enumerate(['axial', 'coronal', 'sagittal']):
        img_path = image_paths[f'{class_name}_{view}']
        img = mpimg.imread(img_path)
        axes[i, j].imshow(img, cmap='gray')
        axes[i, j].axis('off')
        axes[i, j].set_title(f'{class_label} - {view.capitalize()}')
plt.tight_layout()
plt.show()
```
